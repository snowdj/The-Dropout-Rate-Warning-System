```python
#Q1男生女生誰比較愛休學?
array = data.values

bs=0 
gs=0 
br=0.000000 
gr=0.000000 
for i in range(0,999,1): 
    if(array[i:i+1,5:6]=="學士班(日間)"):
        b=array[i:i+1,11:12]
        b1=b.flatten() #[['2969']]-->['2969']
        b2 = ''.join(map(str, b1)) #['2969']-->'2969'(一維才能使用)
        b3=int(b2)
        bs=bs+b3;
        g=array[i:i+1,12:13]
        g1=g.flatten() #[['2969']]-->['2969']
        g2 = ''.join(map(str, g1)) #['2969']-->'2969'(一維才能使用)
        g3=int(g2)
        gs=gs+g3;
br=float(bs)/float(bs+gs) 
gr=float(gs)/float(bs+gs) 
br=br*100 
gr=gr*100 
print("全國男生學士班(日間)休學人數",bs) 
print("全國女生學士班(日間)休學人數",gs) 
print("男生百分比",br) 
print("女生百分比",gr)

for i in range(0,999,1): 
    if(array[i:i+1,5:6]=="學士班(日間)"):
        b=array[i:i+1,8:9]
        b1=b.flatten() #[['2969']]-->['2969']
        b2 = ''.join(map(str, b1)) #['2969']-->'2969'(一維才能使用)
        b3=int(b2)
        bs=bs+b3;
        g=array[i:i+1,9:10]
        g1=g.flatten() #[['2969']]-->['2969']
        g2 = ''.join(map(str, g1)) #['2969']-->'2969'(一維才能使用)
        g3=int(g2)
        gs=gs+g3;
data.head()
```
